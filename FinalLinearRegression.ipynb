```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv("datasets/housedata.csv")
df2 = pd.read_csv("datasets/Nearby_features.csv")
```

```python
joindf = df.join(df2.set_index('PID'), on='PID')
```

```python
joindf = joindf.set_index('PID')
```

```python
joindf
```

```python
joindf = pd.get_dummies(joindf,columns=['Property City', 'House Type','Zone'])
```

```python
joindf = joindf.dropna()
```

```python
joindf.isna().sum()
```

```python
# Import dependencies
from sklearn.linear_model import LinearRegression

# Generate some data
X = joindf.drop(columns = ['Sale Price(USD)', 'Property Address', 'Sale Date'])
y = joindf['Sale Price(USD)']

# Create a linear model
model = LinearRegression()

# Fit (train) our model to the data
model.fit(X, y)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Use our model to predict a value
predicted = model.predict(X)

# Score the prediction with MSE and R2
mse = mean_squared_error(y, predicted)
r2 = r2_score(y, predicted)

print(f"mean squared error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
# Overall Score for the model
model.score(X, y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
coef = model.coef_
coef = list(coef)
coef
```
